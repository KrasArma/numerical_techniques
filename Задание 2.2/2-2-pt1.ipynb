#### Задание 2.2. часть 1
##### Найти область локализации корня для первой точки
###### В математике и численных методах область локализации корня — это интервал или область, в которой находится корень уравнения или системы уравнений. Определение такой области является важным шагом в численных методах, так как оно позволяет использовать различные алгоритмы для нахождения корней, такие как метод бисекции, метод Ньютона и другие. В данном докладе мы рассмотрим задачу нахождения области локализации корня для системы уравнений, состоящей из двух функций.
###### Область локализации корня необходима для oпределения начальных условий, устойчивости методов, оптимизации вычислений
#### Ход решения задачи
###### Импорт зависимостей 

```python
import numpy as np
import matplotlib.pyplot as plt
```

###### Определение вводных


```python
a = 1

def f1(x1, x2):
    return x1 - np.cos(x2) - 1

def f2(x1, x2):
    return x2 - np.log(x1 + 1) - a
```

###### Определение функция для нахождения области локализации корня


```python

def find_root_region(x1_range, x2_range, step=0.1):
    x1_values = np.arange(x1_range[0], x1_range[1], step)
    x2_values = np.arange(x2_range[0], x2_range[1], step)
    
    f1_values = []
    f2_values = []
    
    for x1 in x1_values:
        for x2 in x2_values:
            f1_val = f1(x1, x2)
            f2_val = f2(x1, x2)
            f1_values.append(f1_val)
            f2_values.append(f2_val)
    
    f1_values = np.array(f1_values).reshape(len(x1_values), len(x2_values))
    f2_values = np.array(f2_values).reshape(len(x1_values), len(x2_values))
    
    return f1_values, f2_values, x1_values, x2_values
```

###### Визуализация

```python

x1_range = (0, 2)
x2_range = (0, 2)

f1_values, f2_values, x1_values, x2_values = find_root_region(x1_range, x2_range)


plt.figure(figsize=(12, 6))


plt.subplot(1, 2, 1)
plt.contourf(x1_values, x2_values, f1_values, levels=np.linspace(-2, 2, 50), cmap='RdYlBu')
plt.colorbar()
plt.title('f1(x1, x2)')
plt.xlabel('x1')
plt.ylabel('x2')


plt.subplot(1, 2, 2)
plt.contourf(x1_values, x2_values, f2_values, levels=np.linspace(-2, 2, 50), cmap='RdYlBu')
plt.colorbar()
plt.title('f2(x1, x2)')
plt.xlabel('x1')
plt.ylabel('x2')

plt.tight_layout()
plt.show()
```

###### Определяем области, где функции меняют знак

```python

def find_sign_change_region(f_values, x1_values, x2_values):
    sign_change_points = []
    for i in range(len(x1_values) - 1):
        for j in range(len(x2_values) - 1):
            if (f_values[i, j] * f_values[i + 1, j] < 0) or (f_values[i, j] * f_values[i, j + 1] < 0):
                sign_change_points.append((x1_values[i], x2_values[j]))
    return sign_change_points

sign_change_f1 = find_sign_change_region(f1_values, x1_values, x2_values)
sign_change_f2 = find_sign_change_region(f2_values, x1_values, x2_values)

print("Области изменения знака для f1:")
for point in sign_change_f1:
    print(f"x1 = {point[0]:.2f}, x2 = {point[1]:.2f}")

print("\nОбласти изменения знака для f2:")
for point in sign_change_f2:
    print(f"x1 = {point[0]:.2f}, x2 = {point[1]:.2f}")
```
