##### Pешить методом итераций сделать оценку сходимости для выбранной области
###### Метод итераций (или метод простых итераций) заключается в том, что мы преобразуем систему уравнений в итерационную форму и последовательно обновляем значения переменных.
###### Шаги решения: oпределение функций, итерации метода, oценка сходимости
#### Ход решения задачи
###### Импорт зависимостей 

```python
import numpy as np
```

###### Определяем вводные

```python
a = 1

def f1(x1, x2):
    return np.cos(x2) + 1

def f2(x1, a=1):
    return np.log(x1 + 1) + a
```

###### Определение и применение метода итераций

```python
def iteration_method(x0, tol=1e-6, max_iter=100):
    x = x0
    for i in range(max_iter):
        x_new = np.array([f1(x[0], x[1]), f2(x[0])])
        if np.linalg.norm(x_new - x) < tol:  
            return x_new, i + 1
        x = x_new
    raise ValueError("Метод не сошелся")

x0 = np.array([1.0, 1.0])

solution, iterations = iteration_method(x0)

print(f"Решение: x1 = {solution[0]}, x2 = {solution[1]}, Итерации: {iterations}")
```

###### Оценка сходимости

```python

print(f"Решение: x1 = {solution[0]}, x2 = {solution[1]}, Итерации: {iterations}")

def convergence_analysis(x0, tol=1e-6, max_iter=100):
    x = x0
    convergence_history = []
    for i in range(max_iter):
        x_new = np.array([f1(x[0], x[1]), f2(x[0])])
        convergence_history.append(np.linalg.norm(x_new - x))
        if np.linalg.norm(x_new - x) < tol:
            break
        x = x_new
    return convergence_history

convergence_history = convergence_analysis(x0)
print("\nОценка сходимости:")
for i, error in enumerate(convergence_history):
    print(f"Итерация {i + 1}: Ошибка = {error}")
```
