##### Определим функцию для правой части уравнения, точное решение, параметры, количество шагов, массивы для хранения значений и начальное значение

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return -y * np.cos(x) + 0.5 * np.sin(2 * x)

def exact_solution(x):
    return np.sin(x) - 1 + np.exp(-np.sin(x))

x0 = 0 
y0 = 0 
x_end = 1 
h = 0.1
n = int((x_end - x0) / h) 

x_values = np.arange(x0, x_end + h, h)
y_values = np.zeros(len(x_values))

y_values[0] = y0
```

##### Применяем метод Эйлера, вычисляем значения и погрешность

```python

for i in range(1, len(x_values)):
    y_values[i] = y_values[i - 1] + h * f(x_values[i - 1], y_values[i - 1])

exact_values = exact_solution(x_values)

errors = np.abs(y_values - exact_values)
```

##### Результаты

```python
print("x values:", x_values)
print("Numerical solution:", y_values)
print("Exact solution:", exact_values)
print("Errors:", errors)

plt.plot(x_values, y_values, label='Numerical Solution (Euler)', marker='o')
plt.plot(x_values, exact_values, label='Exact Solution', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Numerical vs Exact Solution')
plt.legend()
plt.grid()
plt.show()
```
