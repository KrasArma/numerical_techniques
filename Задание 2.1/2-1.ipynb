
##### Задание 2.1
###### Методами простой итерации и Ньютона найти положительный корень нелинейного уравнения x^3+x^2-x-0.5=0; начальное приближение определить графически.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**3 + x**2 - x - 0.5

def f_prime(x):
    return 3*x**2 + 2*x - 1
```

```python

def simple_iteration(g, x0, tol=1e-5, max_iter=100):
    x = x0
    for i in range(max_iter):
        x_new = g(x)
        if abs(x_new - x) < tol:
            return x_new, i
        x = x_new
    return x, max_iter

def newton_method(x0, tol=1e-5, max_iter=100):
    x = x0
    for i in range(max_iter):
        x_new = x - f(x) / f_prime(x)
        if abs(x_new - x) < tol:
            return x_new, i
        x = x_new
    return x, max_iter
```

```python

x = np.linspace(-2, 2, 400)
y = f(x)

plt.plot(x, y, label='f(x) = x^3 + x^2 - x - 0.5')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.title('График функции')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.legend()
plt.show()
```

```python

x0 = 0.5

g = lambda x: (0.5 - x**2 + x)**(1/3) 
root_iter, iter_count = simple_iteration(g, x0)
print(f"Корень методом простой итерации: {root_iter}, итераций: {iter_count}")

```

```python
root_newton, newton_count = newton_method(x0)
print(f"Корень методом Ньютона: {root_newton}, итераций: {newton_count}")
```
